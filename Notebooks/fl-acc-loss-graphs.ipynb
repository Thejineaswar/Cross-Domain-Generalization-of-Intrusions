```python
import pandas as pd
import pickle
```

```python
with open("../input/notebooka0ac64623c/test.txt", "rb") as fp:   # Unpickling
    b = pickle.load(fp)
```

```python
data_dict = {
    0: [],
    1: [],
    2: [],
    3: [],
    4: [],
    5: [],
}
```

```python
CLIENT_PRINT = {
    0: "CICIDS 2018",
    1: "CICIDS 2017",
    2:  "BOT IOT",
    3: "NSL_KDD",
    4:"TON_IOT",
    5:"UNSW_NB15"
}
```

```python
for k in range(0,6):
    for i in range(0,99):
        data_dict[k].append(b[k][i][4])    
```

```python
import numpy as np
import plotly.graph_objects as go

num_rounds = [i for i in range(0,99)]

fig = go.Figure()

for i in data_dict.keys():
        temp = data_dict[i]
        fig.add_trace(go.Scatter(
                            x=num_rounds, y=temp,
                            mode='lines',
                            name=CLIENT_PRINT[i]
                        )
            )

fig.update_layout(
    title = 'Accuracy v/s training rounds in Federated Environment',
    xaxis_title = 'Training rounds',
    yaxis_title = 'Accuracy',
    height = 400,
    width = 600
)
fig.show()
```

```python
loss_dict = {
    0: [],
    1: [],
    2: [],
    3: [],
    4: [],
    5: [],
}
```

```python
for k in range(0,6):
    for i in range(0,99):
        loss_dict[k].append(b[k][i][2])    
```

```python
fig = go.Figure()

for i in loss_dict.keys():
        temp = loss_dict[i]
        fig.add_trace(go.Scatter(
                            x=num_rounds, y=temp,
                            mode='lines',
                            name=CLIENT_PRINT[i]
                        )
            )

fig.update_layout(
    title = 'Loss v/s training rounds in Federated Environment',
    xaxis_title = 'Training rounds',
    yaxis_title = 'Loss',
    height = 400,
    width = 600
)
fig.show()
```
